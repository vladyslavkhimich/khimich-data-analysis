```python
import csv

warehouse = [
  {'name': 'Xiaomi', 'number': 180, "price": 13, "type": "budget"},
  {'name': 'Oppo', 'number': 50, "price": 18, "type": "mid-range"},
  {'name': 'Google', 'number': 78, "price": 20, "type": "flagship"},
  {'name': 'Apple', 'number': 670, "price": 25, "type": "flagship"},
]

with open('warehouse.csv', 'wt', newline='') as frecord:
  crecord = csv.DictWriter(frecord, ['name', 'number', 'price', 'type'])
  crecord.writeheader()
  crecord.writerows(warehouse)
```
